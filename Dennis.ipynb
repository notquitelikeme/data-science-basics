```python
# assignment
# pick one dataset of your choice and do either classification, regression or clustering
# What is the Difference?
# classification- discrete values(the outcome is either yes or no) most of the time the outcome is inform of text
# regression- continuous values (the outcome is numerical and continuous) mainly used to predict sales, prices, size
# clustering- cluster items based on similarities shared by members of one cluster
import pandas
data=pandas.read_csv("fruits.csv")
data
```

```python
data.describe()
```

```python
# I will do classification
# splitting data into x and y
array=data.values
X=array[:, 1:5]
Y=array[:, 5]
```

```python
# splitting the data into test data and train data
from sklearn import model_selection
X_test,X_train,Y_test,Y_train=model_selection.train_test_split(X,Y, test_size=0.30,random_state=42)
```

```python
# classification models
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()
# fit the data into the model
model.fit(X_train,Y_train)
```

```python
prediction=model.predict(X_test)
print("The model predicted is", prediction)
print("The model expected is", Y_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(prediction,Y_test))
```
