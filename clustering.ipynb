```python
# classification and regression is part of supervised learning
# clustering is part of unsupervised learning
# we don't have the Y value in clustering
# therefore we do not train the model to predict an outcome
# we only give the model some dataset and the model learns by itself
# after learning, the model can group data into different categories based on similarities in them
# so basically, clustering automatically groups data according to similar objects
# data is grouped or categorized into sets/clusters
import pandas
data=pandas.read_csv(r"Mall_Customers.csv")
data
```

```python

# check if there are null values
data.isnull().sum()
```

```python
# clean dataset
data['Annual Income'].fillna('0',inplace=True)
data['Spending Score'].fillna('0',inplace=True)
```

```python
# step 2
# set the x value from the dataset
array=data.values
x=array[:,2:5]
```

```python
# step 2
# set the x value from the dataset
array=data.values
x=array[:,2:5]
```

```python
# select the algorithm/model to use
from sklearn.cluster import KMeans
# initialize the model
model=KMeans(n_clusters=10,random_state=42)
# fit the data into the model
model.fit(x)
```

```python
# display the clusters(10)
means=model.cluster_centers_
clusters=pandas.DataFrame(means,columns=['Age','Annual Income','Spending Score'])
clusters
```

```python
# view one cluster
data['label']=model.labels_
cluster3=data[data['label']==3]
cluster3
```
