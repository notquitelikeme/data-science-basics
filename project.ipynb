```python
# we are going to do a project on movie recommendation
# the model will learn based on the data of users and movie details
# the model will then calculate the mean of rating by different users
# we will also count the number of users who rated a particular movie
# the recommendation will be based on movies under the same category/cluster of rating
import pandas
# modcom.co.ke/datasets/file.csv
# modcom.co.ke/datasets/Movie_Id_Titles.csv
User_items_details=pandas.read_csv("file.csv")
User_items_details.tail(4) # reading the first records
```

```python
# the second dataset
Movie_details=pandas.read_csv("Movie_Id_Titles.csv")
Movie_details.tail(4)
```

```python
data=pandas.merge(User_items_details,Movie_details)
data.tail(10)
```

```python
# group the dataset based on the rating
# find the mean as per rating
data.groupby('title')['rating'].mean().sort_values(ascending=False).head(50)
```

```python
# how many times the movie was rated/how many users
data.groupby('title')['rating'].count().sort_values(ascending=False).tail(50)
```

```python
# display the mean in a dataframe
mean_rating=pandas.DataFrame(data.groupby('title')['rating'].mean())
# this is for displaying the total count
mean_rating['n_ratings']=pandas.DataFrame(data.groupby('title')['rating'].count())
mean_rating.head(7)
```

```python
# pivot table
# find which movie each user rated
# user, movie, rating of the movie
pivot=data.pivot_table(index='user_id',columns='title',values='rating')
pivot.tail(10)
```

```python
# display movies that have close(high) correlation
# movies that correlate based on the ratings
selected_movie=pivot['Star Wars (1977)']
# finding similar movies based on their ratings
similar=pivot.corrwith(selected_movie)
similar_df=pandas.DataFrame(similar,columns=['Correlations'])
similar_df.sort_values('Correlations',ascending=False).head(50)
```

```python
# show movies rated by more than 200 users/num_of_ratings more than 200
similar_df=similar_df.join(mean_rating['n_ratings'])
similar_df.tail(5)
```

```python
# pick only the movies with more than 200 ratings
similar_df[similar_df['n_ratings']>200].sort_values('Correlations',ascending=False)
```
