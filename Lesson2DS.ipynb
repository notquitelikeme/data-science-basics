```python
# classification
# grouping data into categories
# model training
# recipe for cooking muffins, cupcakes, scones
# we will use some data generated by Alice when she did some research on the amounts of ingredients to use
# at the end of the training, we will give the model some data to test the outcome/ whether we are cooking cupcakes,muffins or scones
# modcom.co.ke/datasets/data.csv
import pandas
data=pandas.read_csv("data.csv")
data
```

```python
data.describe()
```

```python
# split the data into x and y
# X = sugar, flour, butter, eggs, baking powder, vanilla, salt, milk[features/ingredients]
# Y = class/type(cupcakes, scones or muffins)
array=data.values
X=array[:, 0:8]
Y=array[:, 8]
```

```python
# spilt the dataset into training and test data
from sklearn import model_selection
X_test,X_train,Y_test,Y_train=model_selection.train_test_split(X,Y, test_size=0.30,random_state=42)
```

```python
# get the classification model to use
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
# select one model to use
model=GaussianNB()
# fit in the training dataset
# we only use 70% to train the model
model.fit(X_train,Y_train)
```

```python
# give the model some data (30%) to predict
prediction=model.predict(X_test)
print("The model predicted",prediction )
print("The expected outcome is: ",Y_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(prediction,Y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(prediction,Y_test))
from sklearn.metrics import confusion_matrix
print(confusion_matrix(prediction,Y_test))
```
