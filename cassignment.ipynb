```python
# modcom.co.ke/datasets/AirlinesCluster.csv
import pandas
data=pandas.read_csv(r"AirlinesCluster.csv")
data
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data['FlightMiles'].fillna('0',inplace=True)
data['FlightTrans'].fillna('0',inplace=True)
data['DaysSinceEnroll'].fillna('0',inplace=True)
```

```python
array=data.values
x=array[:,2:6]
x
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=10,random_state=42)
model.fit(x)
```

```python
means=model.cluster_centers_
clusters=pandas.DataFrame(means,columns=['QualMiles','FlightMiles','FlightTrans','DaysSinceEnroll'])
clusters
```

```python
data['label']=model.labels_
cluster3=data[data['label']==3]
cluster3
```
